```python
c1 = 'dimgray' ## protein coding genes
c2 = myviol ## centromeric genes
c3 = myblue ## rest of heterochromatic genes
c4 = 'g' ## subtelomeric genes
c5 = 'xkcd:salmon' ## mating type region
```

```python
# Legend for boxplots
fig, ax = plt.subplots(figsize=(48,6))

patch1 = mpatches.Patch(color=c1, label='protein coding')
patch2 = mpatches.Patch(color=c5, label='mat locus')
#patch3 = mpatches.Patch(color=c2, label='$\it{dg+/-}$')
#patch4 = mpatches.Patch(color=c2, label='$\it{dh+/-}$')
patch3 = mlines.Line2D([], [], color=c2, marker='o', markersize=50, markeredgecolor="k", label='$\it{dg+/-}$')
patch4 = mlines.Line2D([], [], color=c2, marker='D', markersize=50, markeredgecolor="k", label='$\it{dh+/-}$')
#patch5 = mpatches.Patch(color=c4, label='$\it{tlh+/-}$')
#patch6 = mpatches.Patch(color=c4, label='$\it{SPAC212.10+/-}$')
patch5 = mlines.Line2D([], [], color=c4, marker='o', markersize=50, markeredgecolor="k", label='$\it{tlh+/-}$')
patch6 = mlines.Line2D([], [], color=c4, marker='D', markersize=50, markeredgecolor="k", label='$\it{SPAC212.10+/-}$')
patch7 = mpatches.Patch(color=c3, label='heterochromatic\ngenes')

ax.axis('off')

all_handles = (patch1, patch2, patch3, patch4, patch5, patch6, patch7)

leg = ax.legend(handles=all_handles, ncol=4, handlelength=.5, handleheight=.5, fontsize=70, 
                loc='center', borderpad=1.5)

frame = leg.get_frame()
frame.set_linewidth(0)

ax.add_artist(leg)
plt.savefig(os.path.join(plots_dir, 'bp_legend.png'))
plt.savefig(os.path.join(plots_dir, 'bp_legend.pdf'), format='pdf', bbox_inches='tight')
plt.show()
```
