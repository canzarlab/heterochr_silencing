```python
in_dir = '/data/parastou/RNAdeg/data/'
```

```python
cols = ['gene-id', 'gene-name', 'biotype', 'desc', 1, 2, 3]
```

```python
df = pd.read_csv(os.path.join(in_dir, 'ribosomal_proteins.txt'), sep=';', quoting = 3, names=cols)
```

```python
df = df[['gene-id', 'gene-name', 'biotype', 'desc']]
```

```python
df['gene-id'] = df['gene-id'].apply(lambda x: x.split('"ID=gene:')[1])
```

```python
df['gene-name'] = df['gene-name'].apply(lambda x: x.split('Name=')[1])
```

```python
ribo_df = df[['gene-id', 'gene-name']]
```

```python
ribo_df.to_csv(os.path.join(in_dir, 'ribosomal.csv'), sep='\t', )
```
